# histograms :D

```python
def getplotdata():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1) 
    
    #i'm guessing we're skipping row 1 because it's just a string?
    
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)
    
    return data
```

```python
plt.style.use('bmh')
data = getplotdata()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=100, label=['ghoti', 'is pronounced like fish'])
plt.xlabel('gonna mess with stuff to see what they do')
plt.ylabel('holy guacamole')
plt.legend()
```

# scatter plots :D

```python
plt.style.use("seaborn-v0_8-darkgrid") #i dont see much difference except the background changed colour
data = getplotdata()

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x,y, label=label, alpha=.5)

plt.xlabel('so i am guessing we make x arbitrarily so it is able to plot?')
plt.ylabel('i wonder why getplotdata takes no argument')
plt.legend()
```
