```python
def renamefile(oldname):
    getdate = os.path.split(oldname)
    datee = getdate[1]

    month = datee[0:3]
    day = datee[4:6]
    extension = os.path.splitext(oldname)
    filetype = extension[1]
    country = datee[7:]
    
    monthno = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
    return(f'{monthno.get(month)}-{day}-{country}')

 
renamefile('os-collaboration-exercise-data\\os-collaboration-exercise-data\\oct-18_tokyo.txt')
#FINALLY I DID IT
```
