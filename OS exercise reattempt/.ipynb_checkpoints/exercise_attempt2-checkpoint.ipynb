```python
def globby(city):
    if city == 'cairo':
        return glob.glob('*/**/*cairo.txt', recursive = True)
    if city == 'dhaka':
        return glob.glob('*/**/*dhaka.txt', recursive = True)
    if city == 'jakarta':
        return glob.glob('*/**/*jakarta.txt', recursive = True)
    if city == 'karachi':
        return glob.glob('*/**/*karachi.txt', recursive = True)
    if city == 'manila':
        return glob.glob('*/**/*manila.txt', recursive = True)
    if city == 'melbourne':
        return glob.glob('*/**/*melbourne.txt', recursive = True)
    if city == 'osaka':
        return glob.glob('*/**/*osaka.txt', recursive = True)
    if city == 'shanghai':
        return glob.glob('*/**/*shanghai.txt', recursive = True)
    if city == 'singapore':
        return glob.glob('*/**/*singapore.txt', recursive = True)
    if city == 'tokyo':
        return glob.glob('*/**/*tokyo.txt', recursive = True)
    else:
        return 'pain, suffering, sobbing, weeping, crying'

globby('dhaka')
```

```python
def movefile(city):
    filees = globby(city)
    try:
        for file in filees: 
            if city == 'cairo':
                shutil.move(file,'places\\Cairo' )
            if city == 'dhaka':
                shutil.move(file,'places\\Dhaka')
            if city == 'jakarta':
                shutil.move(file, 'places\\Jakarta')        
            if city == 'karachi':
                shutil.move(filees,'places\\Karachi')
            if city == 'manila':
                shutil.move(file,'places\\Manila')
            if city == 'melbourne':
                shutil.move(file,'places\\Melbourne')
            if city == 'osaka':
                shutil.move(file, 'places\\Osaka')
            if city == 'shanghai':
                shutil.move(file, 'places\\Shanghai')
            if city == 'singapore':
                shutil.move(file, 'places\\Singapore')
            if city == 'tokyo':
                shutil.move(file, 'places\\Tokyo')
            else:
                return 'pain, suffering, sobbing, weeping, crying'
    except FileExistsError:
        return 'womp womp'
```
