# exercise 1 (fill in the blanks)

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='firebrick', linestyle='dashed',
             marker='.', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors', markersize = 15)
#i give up looking for the exact marker size 15 should work

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# exercise 2 (plotting a not so simple math plot)
## make y = e^(-x/10)sinx (too lazy to format this)

```python
x_values = np.linspace(0,50,1000)
y_values = np.sin(x_values) * np.exp(-x_values/10)

#perhaps i'm looking at an old version of the website because mine says use fontsize 15
plt.xlabel('x', c='darkorange', fontsize = 20)
plt.ylabel('y', c='darkorange', fontsize = 20, rotation=0)
plt.ylim([-1,1])
plt.title('$y = e^{-x/10}sin(x)$')
plt.grid(alpha=.25)
plt.plot(x_values, y_values)
plt.show()
```
