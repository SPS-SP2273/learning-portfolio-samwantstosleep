```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 4.0
#gonna mess with values to see what they do

plt.plot(x, y_1, label='Y1 values', color='lime', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='cyan', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='magenta')
#cool even these obnoxious colours are in matplotlib
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

#this is when errors are constant. errors can be inputted as a list too
```

```python
#numpy supremacy coz can add math stuff
x = np.array([0,1,2,3,4,5,6])
x3 = x**3
sinx = np.sin(x)

plt.plot(x, x3, label = 'helloo', c='salmon')
plt.plot(x, sinx, label = 'yay colours', c='aquamarine')
plt.legend(loc='lower right')
plt.show()
```

### remember linspace() exists

```python
#linspace creates evenly spaced intervals
x = np.linspace(0,6,100) #points from 0 to 6, but 100 of them
xsquare = x**2
sin_x = np.sin(x)
#got errors but solved it by adding those

plt.plot(x,xsquare,c='lightcoral')

plt.twinx() #this command right here

plt.plot(x, sin_x, c='powderblue')

#yay it looks nice

plt.savefig('sinx and xsquare.png', dpi=150) #ooh it saved in the folder
```
